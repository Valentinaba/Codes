```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_excel("Accidentalidad_2016.xlsx")
```

```python
H=df['HORA']=pd.to_datetime(df['HORA'])

H=H.apply(lambda x: x.hour)

#HH=pd.Series.tolist(HH)
```

```python
GRAVEDAD=[]
for i in range (len(df)):
    
    m=df['GRAVEDAD'][i]
    if m=='MUERTO':
        y=100
    elif m=='HERIDO':
        y=50
        
    elif m=='SOLO DAÑOS':
        y=0
    else:
        pass
    i=i+1
    GRAVEDAD.append(y)
#print (Y)
Y1=[]
for i in range (len(df)):
    
    m=df['GRAVEDAD'][i]
    if m=='MUERTO':
        y=300
    elif m=='HERIDO':
        y=20
        
    elif m=='SOLO DAÑOS':
        y=10
    else:
        pass
    i=i+1
    Y1.append(y)
Y2=[]
for i in range (len(df)):
    
    m=df['GRAVEDAD'][i]
    if m=='MUERTO':
        y=500
    elif m=='HERIDO':
        y=30
        
    elif m=='SOLO DAÑOS':
        y=10
    else:
        pass
    i=i+1
    Y2.append(y)
```

```python
#GRavedad Por Horas
df.plot(kind="scatter", x="X", y="Y", alpha=0.5,
s=Y2, label="Gravedad",
c=H, cmap=plt.get_cmap("gray"), colorbar=True, figsize=(18,20),
)
plt.legend()
plt.title('2016')
#plt.savefig('MapaHora3.png')
```

```python
df.plot(kind="scatter", x="X", y="Y", alpha=0.4,
s=Y1, label="Gravedad",
c=GRAVEDAD, cmap=plt.get_cmap("YlOrRd"), colorbar=True, figsize=(18,20),
)
plt.legend()
plt.title('2014')
#plt.savefig('Mapa3.png')
```

```python
df1=pd.read_excel("Accidentalidad_2015.xlsx")
df1['HORA']=pd.to_datetime(df1['HORA']).apply(lambda x: x.hour)
```

```python
acc_per_hour = pd.value_counts(df1['HORA'].values).sort_index()
Y=[]
X=np.arange(0,24,1)
for i in range(len(acc_per_hour) ):
    l=df1[(df1['HORA']==i)]
    l=l[l['GRAVEDAD']=='MUERTO']
    S=(len(l)/acc_per_hour[i])*100
    #print(i,S)
    Y.append(S)
plt.title('Porcentaje de muerte por hora para el 2015')
plt.plot(X,Y,'.')
plt.xlabel('Hora del dia')
plt.ylabel('Porcentaje')
plt.grid(alpha=0.4)
```
